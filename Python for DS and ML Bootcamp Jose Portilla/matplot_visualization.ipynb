```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x ** 2
```

```python
%matplotlib inline
```

```python
plt.plot(x,y)
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.6,0.2,0.3,0.2])

axes1.plot(x,y)
axes1.set_title("Larger Plot")

axes2.plot(y,x)
axes2.set_title("Smaller Plot")
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
# plt.tight_layout()
```

```python
axes
```

Figure Size and DPI

```python
fig = plt.figure(figsize = (3,2))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize = (8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig("figure_save.png",dpi = 200)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_title("Title")
```

```python
# Legend
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x ** 2,label = "X sqrd")
ax.plot(x,x**3,label = "X cubd")

ax.legend(loc = 0) 

"""loc is used to define the location of the legend i.e
    0 = best
    1 = upper right
    2 = upper left
    .
    .
    10 = center
    
    also we can give the 2 tuple like 
    ax.legend(loc = (0.1,0.2)) where 0.1 is the how much distance from the left and 0.2 is how much distance from the bottom ...
    location of the 2 tuple is similar to the one that was done in the subplot section above
"""
```

Part 3 Matplotlib

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = "purple",lw = 1, ls = "-", marker = "1",markersize = 20) # lw = linewidth and ls = linestyle
```
