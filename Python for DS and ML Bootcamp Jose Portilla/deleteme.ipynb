```python
nums = [1,1,0,1,1,1,1,0,1,1,1]
current = 0
m = 0

for i in range(len(nums)):
    if nums[i] == 1:
        current += 1
    else:
        m = max(current,m)
        current = 0
m = max(current,m)
print(m)
```

```python
nums = [1,2,2,4]
duplicate_list = []
for i in range(1,len(nums)+1):
    duplicate_list.append(i)
print(duplicate_list)
```

```python
for i in nums:
    for j in duplicate_list:
        if i != j:
            break
print(i,j)
```
