How many numbers are smaller than the current number? 
Input: nums = [8,1,2,2,3]
Output: [4,0,1,1,3]

```python
nums = [8,1,2,2,3]
output_list = []
count = 0
for i in nums:
    for j in nums:
        if j < i:
            count += 1
    output_list.append(count)
    count = 0
```

```python
print(output_list)
```

Build an Array with Stack operations

target = [1,3], n=3

output = ["Push","push","pop","push"]

```python
target = [1,3]
n = 4
check_list = []
result = []
for i in range(1,n+1):
    if check_list == target:
        break
    check_list.append(i)
    result.append("Push")
    if i not in target:
        check_list.pop()
        result.append("Pop")
        
    
```

```python
print(result)
```

```python
target = [1,3]
n = 4
check_list = []
result = []
k = 0

for i in range(1,n+1):
    if k == len(target):
        break
    
    check_list.append(i)
    result.append("Push")
    
    if i == target[k]:
        k += 1
    else:
        check_list.pop()
        result.append("Pop")

print(result)
    
```

## Reverse Polish Notation

Input: tokens = ["2","1","+","3","*"]

Output: 9

Explanation: ((2 + 1) * 3) = 9

Input: tokens = ["4","13","5","/","+"]

Output: 6

Explanation: (4 + (13 / 5)) = 6

```python
tokens = ["10","6","9","3","+","-11","*","/","*","17","+","5","+"]
calc = []
for token in tokens:
    if token in {"+", "-", "*", "/"}:
        if token == "+":
            calc[-2] = calc[-2] + calc[-1] 
        if token == "-":
            calc[-2] = calc[-2] - calc[-1]
        if token == "*":
            calc[-2] = calc[-2] * calc[-1]
        if token == "/":
            calc[-2] = int(calc[-2] / calc[-1])
        calc.pop()
    else:
        calc.append(int(token))

#print(round(calc[0]))
print(calc[0])
```
