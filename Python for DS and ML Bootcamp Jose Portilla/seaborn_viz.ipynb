# Categorical Plots

```python
sns.barplot(x="sex",y="total_bill",data = tips,estimator=np.std)
```

```python
sns.countplot(x="sex",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
sns.boxplot(x="day",y="total_bill",data=tips,hue="smoker") # we can divide the hue using the categorical column
```

```python
sns.violinplot(x="day",y="total_bill",data = tips,hue="day",legend = False,palette="pastel")
```

```python
sns.stripplot(x="day",y="total_bill",data=tips,hue = "sex")
```

```python
sns.swarmplot(x="day",y="total_bill",data=tips,hue="sex")
```

```python
sns.catplot(x="day",y="total_bill",data=tips,kind="bar") #sns.factorplot was removed
```

```python
numerical_tips
```

```python
numerical_tips.corr()
```

```python
tc = numerical_tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap="coolwarm")
```

## Grids

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris,hue="species",palette="coolwarm")
```

```python
sns.PairGrid(iris) # its a parent of the pairplot which is more customizable.
```

```python
import matplotlib.pyplot as plt

g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

plt.show()
```

```python
g = sns.FacetGrid(data=tips,col="time",row="smoker")
g.map(sns.histplot,"total_bill",kde = True)
```

```python
g = sns.FacetGrid(data=tips,col="time",row="smoker")
g.map(sns.scatterplot,"total_bill","tip")
```
